```python
# create a node
class Node :
    def __init__(self,value):
        self.data=value
        self.next=None
```

```python
# create a empty linked list
class LinkedList:
    def __init__(self):
        # empty LL
        self.head = None
        # count of nodes
        self.n=0
    
    # lets know the length of the LL
    def __len__(self):
        return self.n
    
    # insert elements through head
    def insert(self, value):
        # create a new node
        new_node = Node(value)
        #create connections
        new_node.next = self.head
        self.head=new_node
        # increment the LL length
        self.n += 1
        # print(self.n)
    
    # traversing functionality    
    def traverse(self):
        # traversing starts from head ,so let current be head
        current = self.head
        # use loop to move from one node to next
        while current is not None:
            print(current.data)
            current=current.next
        
    # customizing print function for LL
    def __str__(self):
        #check if the LL is empty
        if self.n ==0:
            return "empty linked list"
        current=self.head
        result=''
        while current is not None:
            result= result+'-->'+str(current.data)
            current=current.next
        return result[:]
    
    # creating append functionality for LL
    def append(self, value):
        #create a new node for the new value
        new_node=Node(value)
        # check if the LL is empty
        if self.head is None:
                self.head = new_node
                self.n+=1
                return
        # traverse or loop to end node
        current=self.head
        while current.next is not None:
            current=current.next
        current.next=new_node
        self.n+=1
    # creating a function to add a new after a value
    def insert_after(self,after,value):
        # create a new node
        new_node=Node(value)
        # check if the LL is empty, if so the make new node the head
        if self.head is None:
            self.head=new_node
            return
        # traverse through LL to find the node after which we want to insert
        current=self.head
        while current.next is not None:
            #check for the item in LL
            if current.data == after:
                break
            current=current.next
        # here LL may not have the item at all,so if current.next is none then the item is'nt present
        if current.next is None:
            print("Item not found")
            return
        else:
            # insert the new node after the current node
            new_node.next=current.next
            current.next=new_node
            self.n+=1
    
    # delete all the node of a LL
    def clear(self):
        self.head=None
        self.n=0
        
    # delete head
    def delete_head(self):
        # check if the LL is empty
        if self.n == 0:
            return "empty link list"
        self.head=self.head.next
        self.n-=1
    
    #delete tail node
    def delete_tail(self):
        # check if the LL is empty
        if self.n == 0:
            return "empty link list"
        # traverse to the second last node
        current=self.head
        while current.next is not None:
            # check if the current node is the second last node
            if current.next.next is None:
                break
            current=current.next
        current.next=None
        self.n-=1
    
    #delete a node after a value
    def delete_middle(self,value):
        #check if the LL is empty
        if self.n == 0:
            return "empty link list"
        
        # if there's only one node
        if self.head.data == value:
            self.head=None
            self.n-=1 
            return      
        # traverse to the second last node before the node to be removed
        current =self.head
        while current.next is not None:
            # check if the current node is the second last node
            if current.next.data == value:
                break
            current=current.next
        # if item is found ,if yes then delete it
        if current.next is None:
            return f'element {value} not found'
        else:
            current.next=current.next.next
            self.n-=1
    
    #find a item position
    def find(self,item):
        #check if the LL is empty
        if self.n == 0:
            return "empty link list"
        #  traverse to the node with the item
        current=self.head
        pos=0
        while current is not None:
            if current.data == item:
                return pos
            current=current.next
            pos+=1
        return "item not found"
    
    # search by index
    def search_indeex(self,index):
        #check if the LL is empty
        if self.n == 0:
            return "empty link list"
        #  traverse to the node at index
        current=self.head
        pos=0
        while current is not None:
            if pos == index:
                return current.data
            current=current.next
            pos+=1
        return "item not found"
    
    #delete by index
    def delete_by_index(self,index):
        #check if the LL is empty
        if self.n == 0:
            return "empty link list"
        #check if the position is 0
        if index == 0:
            self.head=self.head.next
            return
        #  traverse to the node at index
        current=self.head
        pos=1
        while current is not None:
            if pos == index:
                break
            current=current.next
            pos+=1
        #check if the node is found
        if current.next == None:
            return f"indexError: there is no {index} index in tis linked list"
        else:
            current.next=current.next.next
```
